# Gradient descent algorithm

```python
# change the shape of y and x 
y = y.reshape(y.shape[0],1)
x = x.reshape(x.shape[0],1)
```

```python
# add the constant and get the matrix X
X = np.hstack(( x, np.ones(x.shape)))
```

```python
# parameters a and b 
theta = np.random.randn(2,1) 
```

## Modèle 

```python
def modele(X, theta) : 
    return X.dot(theta)
```

```python
modele(X, theta).shape
```

```python
plt.plot(x, y, c = 'red') 
plt.scatter(x, modele(X, theta))
```

```python
## the sum of squares of the residuals 
def residuals(X ,y ,theta) :
    n = len(y)
    return  (1/2*n) *np.sum(( modele(X, theta) - y)**2)
```

## Gradient descent 

```python
def grad(X,y,theta) : 
    n = len(y)
    return (1/n)* X.T.dot(modele(X, theta) -  y) 
```

```python
def grad_descent(X, y , theta, learning_rate, iterations) : 
    
    cost = np.zeros(iterations)
    for i in range (0, iterations) : 
        theta = theta - learning_rate * grad(X,y,theta)
        cost[i] = residuals(X ,y ,theta) 
    return theta,  cost
```

```python
theta_regression, cost = grad_descent(X, y , theta, learning_rate = 0.01, iterations = 1000)
```
